```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.tradingview.com/markets/stocks-usa/market-movers-all-stocks/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('div', class_= "shadow-zuRb9wy5")
```

```python
table = soup.find('div', class_= "shadow-zuRb9wy5")
```

```python
print (table)
```

```python
trading_title = table.find_all('th', class_= "cell-seAzPAHn")
```

```python
trading_title
```

```python
clean_trading_title = [title.text for title in trading_title]
```

```python
data_for_all = pd.DataFrame(columns = clean_trading_title)

data_for_all
```

```python
soup.find_all('table', class_="table-Ngq2xrcG") 
```

```python
row_data = soup.find_all('tr', class_ = "row-RdUXZpkv")
```

```python
row_data
```

```python
for row in row_data[1:]:
    rows_data = row.find_all('td')
    individual_row = [data.text.strip() for data in rows_data]
    length = len(data_for_all)
    data_for_all.loc[length] = individual_row
```
